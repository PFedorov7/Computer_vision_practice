```python
#Convert to PNG to PDF
for infile in filelist:
    if infile.startswith('.'):
        continue
    outfile = os.path.splitext(infile)[0] + ".pdf"
    try:
        rgba = Image.open(infile)
        rgb = Image.new('RGB', rgba.size, (255, 255, 255))  # white background
        rgb.paste(rgba, mask=rgba.split()[3])               # paste using alpha channel as mask
        rgb.save(outfile, 'PDF', resoultion=100.0)
    except IOError:
        print('Невозможно преобразовать')
```

```python
#PCA

from PIL import Image 
from numpy import * 
from matplotlib import pylab
import pca
import os

os.chdir('/Users/pavelfedorov/Desktop/Practice/Computer_vision_practice/Data/a_selected_thumbs')
mypath = os.getcwd()

imlist = []
for (dirpath, dirnames, filenames) in os.walk(mypath):
    imlist.extend(filenames)
    break
    

im = array(Image.open(imlist[0])) # open one image to get size m,n = im.shape[0:2] # get the size of the images
m,n = im.shape[0:2] # get the size of the images
imnbr = len(imlist) # get the number of images
# create matrix to store all flattened images
immatrix = array([array(Image.open(im)).flatten() for im in imlist],'f')
# perform PCA
V,S,immean = pca.pca(immatrix)
# show some images (mean and 7 first modes)
pylab.figure()
pylab.gray()
pylab.subplot(2,4,1) 
pylab.imshow(immean.reshape(m,n)) 
for i in range(7):
    pylab.subplot(2,4,i+2) 
    pylab.imshow(V[i].reshape(m,n))
pylab.show()
```

```python
# Image derivatives

from PIL import Image
from numpy import *
from scipy.ndimage import filters

im = array(Image.open('/Users/pavelfedorov/Desktop/Practice/Computer_vision_practice/Data/empire.jpg').convert('L'))
#Sobel derivative filters
imx = zeros(im.shape) 
filters.sobel(im,1,imx)
imy = zeros(im.shape) 
filters.sobel(im,0,imy)
magnitude = sqrt(imx**2+imy**2)

pil_im = Image.fromarray(uint8(imy))
pylab.imshow(array(pil_im))
pylab.show()
```

```python
sigma = 2 #standard deviation 
imx = zeros(im.shape)
filters.gaussian_filter(im, (sigma,sigma), (0,1), imx)
imy = zeros(im.shape)
filters.gaussian_filter(im, (sigma,sigma), (1,0), imy)

pil_im = Image.fromarray(uint8(imx))
pylab.imshow(array(pil_im))
pylab.show()
```

```python
#Morphology - counting objects

from scipy.ndimage import measurements,morphology
# load image and threshold to make sure it is binary
im = array(Image.open('/Users/pavelfedorov/Desktop/Practice/Computer_vision_practice/Data/empire.jpg').convert('L'))
im = 1*(im<128)
labels, nbr_objects = measurements.label(im) 
print ("Number of objects:", nbr_objects)
pylab.imshow(array(im))
pylab.show()
```

```python
# morphology - opening to separate objects better
im_open = morphology.binary_opening(im,ones((9,5)),iterations=2) 
labels_open, nbr_objects_open = measurements.label(im_open)
print ("Number of objects:", nbr_objects_open)
pylab.imshow(array(labels_open))
pylab.show()
```
