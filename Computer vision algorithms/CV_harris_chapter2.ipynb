```python
#Harris corner detector
from scipy.ndimage import filters
from PIL import Image
from numpy import *
import harris

im = array(Image.open('/Users/pavelfedorov/Desktop/Practice/Computer_vision_practice/Data/empire.jpg').convert('L'))
harrisim = harris.compute_harris_response(im, 5)
filtered_coords = harris.get_harris_points(harrisim, 15)
harris.plot_harris_points(im, filtered_coords)
```

```python
import pydot
g = pydot.Dot(graph_type='graph')
g.add_node(pydot.Node(str(0),fontcolor='transparent')) 
for i in range(5):
    g.add_node(pydot.Node(str(i+1))) 
    g.add_edge(pydot.Edge(str(0),str(i+1))) 
for j in range(5):
    g.add_node(pydot.Node(str(j+1)+'-'+str(i+1)))
    g.add_edge(pydot.Edge(str(j+1)+'-'+str(i+1),str(j+1))) 
```
