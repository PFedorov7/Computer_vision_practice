```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import log_loss
```

```python
data = pd.read_csv('gbm-data.csv')
data_np = data.values
```

```python
X = data_np[:,1:]
y = data_np[:,0]
```

```python
print(data_np)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=241)
```

```python
loss_test = []
loss_train = []
y_to_sigmoid = []
for i in [1, 0.5, 0.3, 0.2, 0.1]:
    clf = GradientBoostingClassifier(n_estimators=250, random_state=241, learning_rate = i)
    clf.fit(X_train, y_train)
        
    #staged_decision_function()
    for iteration, y_pred in enumerate(clf.staged_predict_proba(X_train)):
        sigm_predict_train = list(map(lambda x : 1 / (1 + np.exp(-x[0])) , y_pred))
        loss = log_loss(y_train, sigm_predict_train)
        loss_train.append(loss)
        
    for iteration1, y_pred1 in enumerate(clf.staged_predict_proba(X_test)):
        sigm_predict_test = list(map(lambda x : 1 / (1 + np.exp(-x[0])) , y_pred1))
        loss1 = log_loss(y_test, sigm_predict_test)
        
        if(i == 0.2):
            print('min logloss is {}, iteration {}'.format(loss1, iteration1))
        
        loss_test.append(loss1)
        
    plt.figure()
    plt.plot(loss_train, 'g', linewidth=2)
    plt.plot(loss_test, 'r', linewidth=2)
    plt.legend(['train', 'test'])

    loss_train.clear()
    loss_test.clear()
```
